```python
from pybaseball import batting_stats
import pandas as pd

batting = batting_stats(2023)
batting = batting[['Name','Team','HR','AVG','OBP','SLG','OPS']]
top_hitters = batting.sort_values(by='OPS', ascending=False).head(10)

print("Top 10 Hitters by OPS (2023):")
print(top_hitters)
```

```python
import matplotlib.pyplot as plt

ax = top_hitters.plot(
    x='Name', y='OPS', kind='barh',
    figsize=(10,6), legend=False, title='Top 10 Hitters by OPS (2023)'
)

ax.invert_yaxis()

for i, (ops, name) in enumerate(zip(top_hitters['OPS'], top_hitters['Name'])):
    ax.text(ops+0.01, i, f"{ops:.3f}", va='center')

plt.xlabel("OPS")
plt.ylabel("Player")
plt.tight_layout()
plt.show()
```

```python
top_hitters.to_csv("top_hitters_2023.csv", index=False)
```
